```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df_col=['type_of_meal_plan','room_type_reserved','market_segment_type','booking_status']
```

```python
for i in df_col:
    df[i]=le.fit_transform(df[i])
```

```python
df.head()
```

```python
for col in df:
    print(col+':')
    print(df[col].unique())
```

```python
df2=df.copy()
```

```python
df2.drop(['Booking_ID','arrival_date'],axis=1,inplace=True)
```

```python
x=df2.iloc[:,:16]
```

```python
x.head()
```

```python
y=df2.iloc[:,-1:]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=50,max_depth=15)
```

```python
model=rfc.fit(x_train,y_train.values.ravel())
```

```python
y_pred=rfc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```
